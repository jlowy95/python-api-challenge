```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
from scipy.stats import linregress

# Import API key
from api_keys import weather_api_key

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (CSV)
output_data_file = "output_data/cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

## Generate Cities List

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
len(cities)
```

### Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* Include a print log of each city as it'sbeing processed (with the city number and city name).


```python
# Set base_url to which full API call will be appended
base_url = "http://api.openweathermap.org/data/2.5/weather"

# Add empty columns of desired data to dataframe
weather_data = {"City":[],"Cloudiness":[],"Country":[],"Date":[],"Humidity":[],"Lat":[],"Lng":[],"Max Temp":[],"Wind Speed":[]}

# Create list of lists for "Sets"
city_data = [[] for i in range(int(len(cities)/10) + 1)]
print("Beginning Data Retrieval\n-----------------------------")

# Initialize city indexer
city_i = 0
# For each set in sets...
for set_num in range(len(city_data)):
    # Initialize record each set so we cycle 0-49
    record = 0
    while (record < 50) & (city_i < len(cities)):
        # Create full url for API call
        call_url = base_url + "?q=" + cities[city_i] + "&units=imperial&appid=" + weather_api_key
        # Store API call
        temp_response = requests.get(call_url)
        # Only take data from good API calls and append to dataframe
        if temp_response.status_code == 200:
            weather_data["City"].append(temp_response.json()['name'])
            weather_data["Cloudiness"].append(temp_response.json()['clouds']['all'])
            weather_data["Country"].append(temp_response.json()['sys']['country'])
            weather_data["Date"].append(temp_response.json()['dt'])
            weather_data["Humidity"].append(temp_response.json()['main']['humidity'])
            weather_data["Lat"].append(temp_response.json()['coord']['lat'])
            weather_data["Lng"].append(temp_response.json()['coord']['lon'])
            weather_data["Max Temp"].append(temp_response.json()['main']['temp_max'])
            weather_data["Wind Speed"].append(temp_response.json()['wind']['speed'])
            
            print(f"Processing Record {record} of Set {set_num+1} | {cities[city_i]}")
            record += 1
        else:
            print("City not found.  Skipping...")
        city_i += 1
print("-----------------------------\nData Retrieval Complete\n-----------------------------")
```

### Convert Raw Data to DataFrame
* Export the city data into a .csv.
* Display the DataFrame

```python
# Convert data dictionary to dataframe
weather_df = pd.DataFrame(weather_data)
# Export dataframe as .csv
weather_df.to_csv("../output_data/weather_data.csv")
# Preview column counts to verify we still have at least 500 good data points
weather_df.count()
```

## Linear Regression

```python
# OPTIONAL: Create a function to create Linear Regression plots
def lin_reg_plot (x_values,y_values,xlabel,ylabel,title):
    # Takes 2 series of values, one for the x-axis and one for the y-axis.  Using the linregress function, it will 
    # compute the slope, intercept, rvalue, pvalue, and stderror for the series.  This function will plot the two
    # series against one another in a scatter plot, including the linear regression and the r-value.
    slope, intercept, rvalue, pvalue, stderr = linregress(x_values, y_values)
    line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
    regress_values = [slope*x + intercept for x in x_values]
    fig = plt.scatter(x_values,y_values,marker="o",color="steelblue")
    plt.plot(x_values,regress_values,"r-")
    plt.annotate(line_eq,(x_values.mean()-10,y_values.mean()-10),fontsize=15,color="red")
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.title(title)
    print(f"The r-squared is: {rvalue}")
    plt.show()
    return fig
```
